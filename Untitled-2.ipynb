```python
import pandas as pd
import os

# List all CSV files in the directory
directory_path = r'C:\Users\CASA\Downloads\arquivos_stream_platforms'
csv_files = [f for f in os.listdir(directory_path) if f.endswith('.csv')]

# Initialize an empty list to store DataFrames
dataframes = []

# Read each CSV file and append to the list
for file in csv_files:
    file_path = os.path.join(directory_path, file)  # Corrected line
    df = pd.read_csv(file_path, sep=',')
    dataframes.append(df)

# Concatenate all DataFrames
combined_df = pd.concat(dataframes, ignore_index=True)

# Display the DataFrame
print(combined_df)

```

```python
# Count rows where "title" column is not empty or null
title_count = combined_df['title'].notna().count()

# Display the count
print(f"Number of rows with a non-empty 'title': {title_count}")
```

```python
# Check if the "genres" column exists
if 'genres' in combined_df.columns:
    # Split the "genres" column into multiple columns
    genres_split = combined_df['genres'].str.split(',', expand=True)
    
    # Add the split columns back to the DataFrame
    combined_df = combined_df.join(genres_split)
else:
    print("The 'genres' column does not exist in one or more files.")##
```
